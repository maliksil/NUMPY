```python
# array: which is used to create data in the form of dimension .

import numpy as np

a=np.array([1,2,3,4,5])
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a[2])
```

```python
# 2D array

a=np.array([[10,21],[9,18]])
print(a)
print(type(a))
print('-----')
print(a.ndim)
print(a.shape)
print(a.size)
print(a[1][0])
```

```python
# arrange : will generate value in 1D.

a=np.arange(1,11)
print(a)
print(a.ndim)
```

```python
# convert  data 1D to 2D and 3D.

a=np.arange(1,61)
print(a)
print(a.shape) # used to find rows and coulmn .
print(a.ndim)  # used to find how many dimension
print(f'total value is :', a.size)  # used to find total how many value is there .
print('------')

b=a.reshape(10,6)  #10 rows and 6 coulmn
print(b)
print(b.ndim)

print(f'index value is :',b[4][3])
```

```python
# convert 1D to 3D

a=np.arange(1,61)
print(a)
print('------')

b=a.reshape(5,2,6)
print(b)

print(f'index value is :',b[2][1][3])
```

```python
# reshape (-1,1)

a=np.array([10,20,30,40,50])
print(a)
print('--------')

b=a.reshape(-1,1)
print(b)

print(b.ndim)
print(b.shape)
```

```python
# reshape (1,-1)

a=np.array([10,20,30,40])
print(a)
print(a.shape)
print('----------')


b=a.reshape(1,-1)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# randn
import numpy as np

a=np.random.randn(3,4)
print(a)
print(type(a))
```

```python
#randint

a=np.random.randint(10,50)
print(a)
print(type(a))
```

```python
#rand

a=np.random.rand(3,4)
print(a)

print(type(a))
```

```python
#random

a=np.random.random(10)
print(a)

print(type(a))
```

```python
# zeros :  default data type of the numpy is float64.

a=np.zeros(3)
print(a)
```

```python
#full

a=np.full((3,4),8)
print(a)

print(type(a))
```

```python
#eye

a=np.eye(3,3)
print(a)
```

```python
# daigonal : it is used to find the daigonals.

import numpy as np

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])

print(a)
print('-------')
print(np.diagonal(a))

```

```python
# ravel : it is a function will convert n-dimension data into 1D.

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
print(a)
print('-------')

b=a.ravel()
print(b)
print(b.shape)
```

```python
#transpose: will convert rows to column and column to rows .

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
print(a)
print('-------')

b=a.T
print(b)
```

```python
#concatination

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
b=np.array([[10,20,30],[1,2,3],[40,50,60]])

print(a)
print('------')

print(b)
print('------')

c=np.concatenate((a,b))

print(c)
```

```python
# axis=1 and axis=0
# in numpy axis=1  : rows and axis=0  : coulmn

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
print(a)
print('------')

print(np.max(a,axis=1))

print('-------')

print(np.min(a,axis=1))

print('--------')

print(np.max(a,axis=1)[0])
```

```python
a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
print(a)
print('------')

print(np.max(a,axis=0))

print('-------')

print(np.min(a,axis=0))

print('--------')

print(np.max(a,axis=0)[0])
```

```python
# argmax and argmin

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
print(a)
print('------')

print(np.argmax(a))

print(np.argmin(a))

```

```python
# matmul

a=np.array ([[10,1,9],[2,8,4],[7,6,13]])
b=np.array ([4,3,2])

print(np.matmul(a,b))
```

```python
# expand_dimension

a=np.array([10,20,30,40,50])
print(a.shape)
print('-------')

b=np.expand_dims(a,axis=0)

print(b)
print(b.shape)
```
